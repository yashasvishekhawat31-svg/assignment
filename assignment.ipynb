Ques 6.What is the difference between import and from ... import in Python?

Ques 7.How can you handle multiple exceptions in Python?

Ques 8.What is the purpose of the with statement when handling files in Python?
Ans.The with statement in Python is used to handle files safely and efficiently by automatically managing resources. When you open a file using with, Python ensures the file is closed automatically, even if an error occurs inside the block. This prevents resource leaks, makes code cleaner, and eliminates the need to manually call file.close(), making file handling more reliable and easier to read.

Ques 9.What is the difference between multithreading and multiprocessing?
Ans.Multithreading uses multiple threads within the same process, sharing the same memory space, making it useful for tasks that involve waiting (like I/O operations), but it is limited by Python’s GIL for CPU-heavy tasks. Multiprocessing uses multiple separate processes, each with its own memory space, allowing true parallel execution and making it better for CPU-intensive tasks, though it requires more memory and overhead than multithreading.

Ques 10.What are the advantages of using logging in a program?

Ques 11.What is memory management in Python?
Ans.Memory management in Python refers to how Python automatically allocates, manages, and frees memory while a program runs. It uses a private heap to store objects, a built-in garbage collector to remove unused objects, and reference counting to track how many variables point to each object. This automatic system ensures efficient use of memory without requiring the programmer to manually allocate or deallocate it.

Ques 12.What are the basic steps involved in exception handling in Python?
Ans.The basic steps in exception handling in Python are: place the risky code inside a try block, handle specific errors using one or more except blocks, optionally run code when no exception occurs using else, and execute cleanup actions that must run regardless of errors using finally.

Ques 13.Why is memory management important in Python?
Ans.Memory management is important in Python because it ensures efficient use of system memory, prevents programs from consuming unnecessary resources, and avoids issues like memory leaks or crashes. Proper memory handling keeps programs running smoothly, improves performance, and allows Python to automatically free unused objects so developers don’t have to manage memory manually.

Ques 14.What is the role of try and except in exception handling?

Ques 15.How does Python's garbage collection system work?
Ans.Python’s garbage collection system automatically manages memory using two mechanisms: reference counting, which immediately frees objects when their reference count drops to zero, and a cyclic garbage collector, which periodically identifies and removes groups of objects involved in circular references that are no longer in use. Together, these processes ensure efficient memory usage without requiring developers to manually free memory.

Ques 16.What is the purpose of the else block in exception handling?

Ques 23.What are the challenges associated with memory management in Python?
Ans.Challenges in Python’s memory management include handling circular references, which can delay garbage collection; dealing with the Global Interpreter Lock (GIL), which can limit true parallelism and affect memory efficiency in multithreaded programs; potential memory leaks caused by lingering references or poorly managed objects; and performance overhead from the garbage collector, which may slow down programs when managing large amounts of data or many objects.

```python
#Ques 6.Write a Python program that logs an error message to a log file when a division by zero exception occurs?
import logging
# Configure logging
logging.basicConfig(filename="errors.log",
                    level=logging.ERROR,
                    format="%(asctime)s - %(levelname)s - %(message)s")
try:
    result = 10 / 0
except ZeroDivisionError:
    logging.error("Division by zero error occurred!")
print("Program finished. Check errors.log for the logged error.")
```

```python
#Ques 18.How would you implement a basic logging setup that logs to a file with rotation after 1MB?
import logging
from logging.handlers import RotatingFileHandler
logger = logging.getLogger("MyLogger")
logger.setLevel(logging.INFO)
handler = RotatingFileHandler(
    "app.log",
    maxBytes=1 * 1024 * 1024,  # 1 MB
    backupCount=3
)
formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")
handler.setFormatter(formatter)
logger.addHandler(handler)
logger.info("Application started.")
logger.error("An error occurred.")
```

```python
#Ques 21. Write a Python program that reads a file and prints the number of occurrences of a specific word?
filename = "example.txt"
word_to_count = "Python"
try:
    with open(filename, "r") as file:
        content = file.read().lower()
        count = content.split().count(word_to_count.lower())
    print(f"The word '{word_to_count}' occurs {count} times in the file.")
except FileNotFoundError:
    print(f"Error: The file '{filename}' does not exist.")
```

```python
#Ques 22.How can you check if a file is empty before attempting to read its contents?
import os
filename = "example.txt"
if os.path.exists(filename):
    if os.path.getsize(filename) > 0:
        with open(filename, "r") as file:
            content = file.read()
            print(content)
    else:
        print("The file is empty.")
else:
    print("Error: The file does not exist.")
```
